```python
#Number of listings in each house type category
df["house_type_id"].value_counts()
```

```python
#Count of properties with unrealistic built sizes (below 20 m² or above 500 m²)
df[(df["sq_mt_built"]<20) | (df["sq_mt_built"]>500)]
```

```python
#Identification of listings with negative or invalid rent values
df=df[df["rent_price"]>=0].reset_index(drop=True)
df[df["rent_price"]<0]
df
```

```python
#Distribution of listings by number of rooms
plt.hist(df["n_rooms"])
plt.show()
```

```python
#Detection of extreme outliers in the buy_price field
q1=df["buy_price"].quantile(0.25)
q3=df["buy_price"].quantile(0.75)
iqr=q3 - q1
upper=q3+1.5*iqr
#lower=q1-1.5*iqr
df[df["buy_price"]>upper].head(10)
```

```python
#Number of listings per built year grouped by decade
df["built_year"]=(df["built_year"]//10)*10
```

```python
df["built_year"].value_counts().sort_index()
```

```python
#Average sale price by built year decade
df.groupby("built_year")["buy_price"].mean()
```

```python
#Count of listings in each neighborhood grouped by district
df.groupby("house_type_id")["neighborhood_id"].nunique().head()
```

```python
#Top twenty streets with the highest number of listings
#Average sale price by floor level
df.groupby("floor")["buy_price"].mean().sort_values(ascending=False)
```

```python
#Identification of duplicated listings based on title
df[df.duplicated('title',keep=False)].sort_values('title').head()
```
